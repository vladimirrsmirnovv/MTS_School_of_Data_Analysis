**Задача №1**. **Сумма первых n положительных чисел**

Напишите программу, запрашивающую у пользователя число и подсчитывающую сумму натуральных положительных чисел от 1 до введенного пользователем значения. Сумма первых n положительных чисел может быть рассчитана по формуле:

Sum=(n)(n+1)/2.

IN: 5

OUT: 15

IN: 0

OUT: 0


```python
def num_sum(num):
    if num < 0:
        print("Введенное число отрицательное! Введите неотрицательное число!")
    else:
        print(f"Введенное число: {num}")
        print(f"Сумма натуральных положительных чисел от 1 до {num} равна {num * (num + 1) // 2}")

num = int(input("Введите число:")) # int уже дает ограничение на то, что пользователь вводит число
num_sum(num)
```

**Задача №2**. **Сумма чисел в числе**

Разработайте программу, запрашивающую у пользователя целое четырехзначное число и подсчитывающую сумму составляющих его цифр. Например, пользователь ввел число 3141, программа должна вывести следующий результат: 3+1+4+1 = 9

IN: 1234

OUT: 10

```python
def sum_of_num(num):
    if 999 < num < 10000:
        first_num = num // 1000 # 3
        second_num = num // 100 % 10 # 1
        third_num = num % 1000 // 10 % 10 # 4
        forth_num = num % 1000 % 10 % 10 # 1
        num_sum = int(first_num + second_num + third_num + forth_num)
        print(num_sum)
    else:
        print(f"Не целое четырехзначное число;\nВведите целое четырехзначное число!")

num = int(input()) # int уже дает ограничение на то, что пользователь вводит число
sum_of_num(num)
```

**Задача №4**. **Високосный год?**

В большинстве случаев год насчитывает 365 дней. Но на самом деле нашей планете требуется чуть больше времени, чтобы полностью пройти по своей орбите вокруг Солнца. В результате для компенсации этой разницы был введен дополнительный день в феврале для особых годов, называемых високосными. Определить, високосный год или нет, можно следуя алгоритму:

*   если год делится на 400 без остатка, он високосный;
*   если год (из оставшихся) делится на 100 без остатка, он НЕ високосный;
*   если год (из оставшихся) делится на 4 без остатка, он високосный;
*   все остальные года не являются високосными.

Напишите программу, запрашивающую год у пользователя и выводящую сообщение о том, високосный ли он.

IN: 2024

OUT: "Високосный"

IN: 2015

OUT: "Не високосный"


```python
year_num = int(input())
if year_num >= 0:
    if year_num % 400 == 0 or (year_num % 100 != 0 and year_num % 4 == 0):
        print("Високосный")
    else:
        print("НЕ високосный")
else:
    print("Введите корректный год!")
```

**Задача №5. Палиндром или нет?**

Строка называется палиндромом, если она пишется одинаково в обоих направлениях. Например, палиндромами в английском языке являются слова «anna», «civic», «level», «hannah». Напишите программу, запрашивающую у пользователя строку и определяющую, является ли она палиндромом.

IN: anna

OUT: "Палиндром"

IN: funny

OUT: "Не палиндром"


```python
def ispalindrome(string):
    string = string.lower()
    if string[::-1] == string:
        print("Введенная строка является палиндромом")
    else:
        print("Введенная строка НЕ является палиндромом")

string = str(input("Введите строку"))
ispalindrome(string)
```

**Задача №7. Вычисляем длину гипотенузы**

Напишите главную функцию, принимающую на вход длины двух катетов прямоугольного треугольника и возвращающую длину гипотенузы, и вспомогательную функцию, рассчитывающую длину гипотенузы по теореме Пифагора. В главной программе должен осуществляется запрос длин сторон у пользователя, вызов вспомогательной функции и вывод на экран полученного результата.

IN:

3

4

OUT:

5

```python
def hypotenuse(a, b):
    return round((a**2 + b**2)**0.5, 4)
def main_hypotenuse():
    a = float(input("Введите длину первого катета: "))
    b = float(input("Введите длину второго катета: "))
    if a > 0 and b > 0:
        result = hypotenuse(a, b)
        print(f"Длина гипотенузы: {result}")
    else:
        print("Введите положительные числа!")

main_hypotenuse()
```

**Задача №8. Простое число?**

Простое число представляет собой число, больше 1, которое без остатка делится лишь на само себя и на единицу. Напишите функцию для определения того. Является ли введенное число простым. Возвращаемое значение должно быть либо True, либо False. В основной программе, как и ожидается, пользователь должен ввести целое число и получить ответ о том, является ли оно простым.

IN: 23

OUT: True

IN: 4

OUT: False

```python
def simple_num(num):  
    if num <= 1:
        return False
    for i in range(2, num):
        if num % i == 0:
            return False
    return True

num = int(input("Введите число: "))
simple_num(num)
```

**Задача №9. Порядок сортировки**

Напишите программу. Которая будет запрашивать у пользователя целочисленные значения и сохранять их в виде списка. Индикатором окончания ввода значений должен служить ноль. Затем программа должна вывести на экран все введенные пользователем числа (кроме нуля) в порядке возрастания – по одному значению в строке. Используйте для сортировки либо метод sort, либо функцию sorted.

IN:

1

5

2

4

0

OUT:

1

2

4

5

```python
my_list = []
while True:
    num = int(input())
    if num == 0:
        break
    else:
        my_list.append(num)
my_list.sort()

for num in my_list:
    print(num)
```

**Задача №10. Избавляемся от дубликатов**

В данном упражнении вам предстоит разработать программу, в которой у пользователя будет запрошен список слов, пока он не оставит строку ввода пустой. После этого на экране должны быть показаны слова, введенные пользователем, но без повторов, - каждое по одному разу. При этом слова должны быть отображены в том же порядке, в каком их вводили с клавиатуры. Например. Если пользователь на запрос программы введет следующий список слов:

IN:

first

second

first

third

second

""

OUT:

first

second

third

```python
my_list_str = []
while True:
    string = str(input())
    if string == "":
        break
    if string not in my_list_str:
        my_list_str.append(string)

for string in my_list_str:
    print(string)
```
