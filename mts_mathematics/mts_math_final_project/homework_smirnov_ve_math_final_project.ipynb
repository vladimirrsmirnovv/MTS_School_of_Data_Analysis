```python
# Функция проверки нормальности распределения
def check_normality(df, column_name):
    values = df[column_name].dropna().values
    stat, p = shapiro(values)
    if p >= 0.05:
        print(f"{column_name}: Данные распределены нормально (p = {p:.5f})")
    else:
        print(f"{column_name}: Данные не распределены нормально (p = {p:.5f})")

# Функция проверки нормальности средних значений по группам
def check_normality_of_means(df, group_column, value_column):
    group_means = df.groupby(group_column)[value_column].mean().values
    stat, p = shapiro(group_means)
    if p > 0.05:
        print(f"Средние значения {value_column} распределены нормально (p = {p:.5f})")
    else:
        print(f"Средние значения {value_column} не распределены нормально (p = {p:.5f})")

# Выбираем количественные переменные для анализа
quantitative_columns = ["iq", "kww", "wage76", "ed76"]

# Проверка нормальности распределений
for col in quantitative_columns:
    check_normality(df, col)

# Проверка нормальности средних значений в разрезе переменной black
group_column = "black"
for col in quantitative_columns:
    check_normality_of_means(df, group_column, col)
```

$H_0$ - выборки уровня образования (`ed76`) по выборкам респонедтов по различным расовым категориям (принадлежностям) имеют одинаковые медианные значения.

$H_1$ - выборки уровня образования (`ed76`) по выборкам респонедтов по различным расовым категориям (принадлежностям) имеют разные медианные значения.

$H_0: {Med}(X_1 - Y_1) = 0$,

$H_1: {Med}(X_1 - Y_1) \neq 0$

```python
# Частоты для каждой категории

black_sample_ed = black_sample['ed76']
white_sample_ed = white_sample['ed76']

black_sample_ed_counts = black_sample_ed.value_counts().sort_index()
white_sample_counts = white_sample_ed.value_counts().sort_index()

# Формируем таблицу сопряженности
contingency_table = pd.DataFrame({
    'Black': black_sample_ed_counts,
    'White': white_sample_counts
}).fillna(0)
```

```python
np.random.seed(42)

def bootstrap_median(sample1, sample2):
    bootstrap_median_diff = [] 
    for _ in range(10000):
        sample1_resampled = np.random.choice(sample1, size=len(sample1), replace=True)
        sample2_resampled = np.random.choice(sample2, size=len(sample2), replace=True)
        median_diff = np.median(sample1_resampled) - np.median(sample2_resampled)
        bootstrap_median_diff.append(median_diff)
    return bootstrap_median_diff


boot_median4 = bootstrap_median(black_sample['ed76'], white_sample['ed76'])

big_ci = np.percentile(boot_median4, q=[2.5, 97.5]).round(2)
print(f"95% доверительный интервал для boot_median4 {big_ci}")


# Тест Пирсона
print(stats.chi2_contingency(contingency_table.T))

# Применяем тест Муда
print(stats.median_test(black_sample['ed76'], white_sample['ed76']))
```

### Точный доверительный интервал для среднего значения (на базе t-критического значения)

$$
\bar x \pm t_{1 - \frac{\alpha}{2}} \cdot \sqrt{\frac{\hat s^2}{n}}.
$$

Так как точные оценки релеватны больше для среднего, то для него (точнее, разницы средних) и оценю (проще моделировать зависимость; в выборке, тем более, крайне серьезных расхождений между медианой и средним не обнаружил).

```python
def t_confidence_interval_mean_diff(sample1, sample2, alpha=0.05):
    n1, n2 = len(sample1), len(sample2)
    mean1, mean2 = np.mean(sample1), np.mean(sample2)
    var1, var2 = np.var(sample1, ddof=1), np.var(sample2, ddof=1)
    se_diff = np.sqrt(var1 / n1 + var2 / n2)
    df = ((var1 / n1 + var2 / n2) ** 2) / ((var1**2 / (n1**2 * (n1 - 1))) + (var2**2 / (n2**2 * (n2 - 1))))
    t_crit = stats.t.ppf(1 - alpha / 2, df)
  
    # Доверительный интервал
    diff_mean = mean1 - mean2
    ci_low = diff_mean - t_crit * se_diff
    ci_high = diff_mean + t_crit * se_diff

    return ci_low, ci_high

ci_low, ci_high = t_confidence_interval_mean_diff(black_sample['iq'], white_sample['iq'])

print(f"95% точный доверительный интервал для разницы средних: [{ci_low:.4f}, {ci_high:.4f}]")

ci_low2, ci_high2 = t_confidence_interval_mean_diff(black_sample['kww'], white_sample['kww'])

print(f"95% точный доверительный интервал для разницы средних: [{ci_low2:.4f}, {ci_high2:.4f}]")


ci_low3, ci_high3 = t_confidence_interval_mean_diff(black_sample['wage76'], white_sample['wage76'])

print(f"95% точный доверительный интервал для разницы средних: [{ci_low3:.4f}, {ci_high3:.4f}]")


ci_low4, ci_high4 = t_confidence_interval_mean_diff(black_sample['ed76'], white_sample['ed76'])

print(f"95% точный доверительный интервал для разницы средних: [{ci_low4:.4f}, {ci_high4:.4f}]")
```

#### Предпосылки Гаусса Маркова

**Предпосылка ГМ-1. Линейная взаимосвязь**

В модели генеральной совокупности зависимая случайная величина $y$ и 
независимая случайная величина $x$ имеют линейную взаимосвязь:
$$
y = \beta_0 + \beta_1 x_1 + \dots + \beta_k x_k + \varepsilon
$$
**Предпосылка ГМ-2. Случайная выборка**

Выборки случайных величин $y$ и $x$, на основании которых строится модель, 
являются репрезентативными случайными выборками своих генеральных 
совокупностей (Independent and Identically Distributed, i.i.d.).

**Предпосылка ГМ-3. Вариативность регрессоров**

**Парная регрессия:** 

Выборка значений случайной величины $x$, а именно $\{x_i, i =1,\dots,n\}$, 
имеет вариативность.

**Множественная регрессия:** 

Отсутствие полной коллинеарности в регрессорах. Выборка значений случайных величин $x_1, \dots, x_k$, а именно $\{x_{i1},\dots,x_{ik}, i =1,\dots,n\}$, имеет вариативность, а также среди регрессоров отсутствует строгая линейная зависимость.

**Предпосылка ГМ-4. Независимость регрессоров и ошибки модели**

Условное математическое ожидание ошибки модели $\varepsilon$ по регрессорам $x_i$ равно 0:

$$
E[\varepsilon \mid x_1, x_2, \dots, x_k] = 0
$$
**Предпосылка ГМ-5. Гомоскедастичность**

Стандартное отклонение ошибки модели $\varepsilon$ постоянно и не зависит от 
регрессоров $x_i$:

$$
\text{Var}(\varepsilon \mid x_1, x_2, \dots, x_k) = \sigma^2
$$

**Предпосылка ГМ-6: Нормальность ошибки**

В генеральной совокупности случайная ошибка $\varepsilon$ имеет нормальное распределение с параметрами:
$$ \varepsilon \sim N(0, \sigma^2) $$

**Что мы имеем теперь?**

Предпосылка №6 не является частью предпосылок Гаусса-Маркова, это дополнительное ограничение, которое мы вводим из-за следующих причин:

- Мы хотим иметь какое-то знание о распределении оценок МНК.
- Ошибка модели аккумулирует в себе эффекты от всех прочих факторов на $y$, которые не включены в уравнение.

Однако, есть подходы для численной оценки уровня мультиколлинеарности. Некоторые из них:

- **Variance Inflation Factor (VIF)** — наиболее часто используемая метрика. Это частная мера зависимости дисперсии оценки коэффициента $\beta_j$ от корреляции рассматриваемого регрессора со всеми прочими регрессорами:
$$\large
VIF_j = \frac{1}{1 - R^2_j}
$$
Откуда получаем, что дисперсию коэффициента можно выразить как:
$$
\text{Var}(\beta_j) = \frac{\sum_{i=1}^{n} r^2_{ij} \varepsilon^2_i}{\sum_{i=1}^{n} (x_{ij} - \bar{x}_j)^2}
$$
**Бенчмарк:**

Если $VIF_j > 10$ (что соответствует $R^2_j > 0.9$), это считается критичным значением, при котором стоит задуматься о работе с факторами.

## Регуляризованная Линейная Регрессия

#### Регуляризация

**Идея:** а давайте штрафовать нашу функцию потерь за то, что в модели используются регрессоры, которые имеют самые большие относительные коэффициенты? Так результаты регрессии станут более стабильными.

Наиболее распространены три вида штрафов:  

- **Least Absolute Shrinkage and Selection Operator (L1 регуляризация)** – сумма модулей 
$$
\large \min \sum (y - \hat{y})^2 + \lambda \sum |\beta_i|
$$
- **Ridge регрессия (Гребневая регрессия, L2 регуляризация)** – сумма квадратов  
$$
\large \min \sum (y - \hat{y})^2 + \lambda \sum \beta_i^2
$$
- **Elastic net** – комбинированная сумма  
$$
\large \min \sum (y - \hat{y})^2 + \lambda_1 \sum |\beta_i| + \lambda_2 \sum \beta_i^2
$$

Lasso фактически отбирает признаки в модель. Давайте попробуем регуляризовать нашу исходную задачу и подобрать коэффициент регуляризации.

```python
# СТАНДАРТИЗАЦИЯ
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# ПОИСК LASSO С ЛУЧШИМ ALPHA
lasso = Lasso(max_iter=5000) 
# Перебор 50 значений alpha
param_grid = {'alpha': np.logspace(-4, 1, 50)}

random_search = RandomizedSearchCV(lasso, 
                                   param_distributions=param_grid, 
                                   n_iter=10, 
                                   cv=5, 
                                   scoring='neg_mean_squared_error', 
                                   random_state=42)
random_search.fit(X_scaled, Y)

# ЛУЧШАЯ МОДЕЛЬ
best_lasso = random_search.best_estimator_
print(f"Лучшее alpha: {random_search.best_params_['alpha']}")

# ВЫВОДИМ КОЭФФИЦИЕНТЫ
lasso_coef = pd.Series(best_lasso.coef_, index=X.columns)
print("\nКоэффициенты модели:")
print(lasso_coef)

# КАКИЕ ПРИЗНАКИ ОБНУЛИЛИСЬ?
print("\nЗануленные признаки:")
print(lasso_coef[lasso_coef == 0].index.tolist())
```

```python
def print_lasso_summary(model, X, Y, best_alpha, feature_names):
    Y_pred = model.predict(X)
    coefficients = pd.Series(model.coef_, index=feature_names)

    r2 = r2_score(Y, Y_pred)
    n = X.shape[0]  # Число наблюдений
    p = X.shape[1]  # Число предикторов
    r2_adj = 1 - ((1 - r2) * (n - 1) / (n - p - 1))
    
    print('\n ========================= TRAIN RESULTS ========================= \n')
    print("=" * 50)
    print(f"Лучшее значение alpha: {best_alpha}")
    print("=" * 50)

    print("\nКоэффициенты модели:")
    print(coefficients.to_string())

    print("\nОценка качества модели:")
    print(f"MSE: {mean_squared_error(Y, Y_pred):.4f}")
    print(f"R²: {r2_score(Y, Y_pred):.4f}")
    print(f"Adjusted R²: {r2_adj:.4f}")

    zero_coef_features = coefficients[coefficients == 0].index.tolist()
    print("\nИсключенные признаки (Lasso занулил их):")
    print(zero_coef_features if zero_coef_features else "Нет исключенных признаков")

# Использование функции
print_lasso_summary(best_lasso, X_scaled, Y, random_search.best_params_['alpha'], X.columns)
```

### Выводы:
- Регуляризация негативно повлияла на $R^2$ (так как мы исключили один из признаков);
- Но и на скорректированный $R^2$ регуляризация повлияла негативно.
